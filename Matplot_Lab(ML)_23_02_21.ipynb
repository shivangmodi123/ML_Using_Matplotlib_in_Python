```
import matplotlib.pyplot as plt
import numpy as np

plt.title("My Graph")

#x=[2,4,6,8]
#y=[4,16,36,64]

t = np.arange(0., 5., 0.2)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^', linewidth=2, linestyle='-')

plt.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd

gas = pd.read_csv('gas_price.csv')

plt.figure(figsize=(20,5))

plt.title('Gas Prize Over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize':18})
plt.plot(gas.Year, gas.USA, 'b--', label='United States')
plt.show()
```
